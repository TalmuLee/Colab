```python
#@title 導入 ETF名詞英翻中
def Category2TWCategory(Category):
  if pd.isna(Category) :
    Category = Category
  EN_list = Category_tables['EN'].to_list()
  TW_list = Category_tables['TW'].to_list()
  if Category in EN_list :
    index = EN_list.index(Category)
    TW_Stxt = TW_list[index]
  else:
    TW_Stxt = Category

  return TW_Stxt
```

```python
#@title 載入台灣時間
def 台灣時間():
  # 取得主機時間(UTC)
  current_time = datetime.datetime.strptime(datetime.datetime.now().strftime("%Y-%m-%d,%H:%M:%S"), "%Y-%m-%d,%H:%M:%S")
  # 加8小時為台灣時區
  new_time = current_time + datetime.timedelta(hours=8)
  # 把時間轉成文字顯示
  Y_M_string = new_time.strftime('%Y%m')
  day_string = new_time.strftime('%m%d')
  time_string = new_time.strftime("%Y-%m-%d,%H:%M:%S")
  Hour_string = new_time.strftime("%H")
  Minute_string = new_time.strftime("%M")
  date_string = datetime.datetime.now().strftime('%Y%m%d')
  time_data_list = [Y_M_string,day_string,time_string,Hour_string,Minute_string,date_string]

  return time_data_list
```

```python
#@title DMA
def DMA(close, weighted):
    # 使用 reduce 計算DMA
    #dma_result = reduce(lambda x, y: (1 - y.iloc[1]) * x + y.iloc[1] * y.iloc[0], zip(close, weighted), close.iloc[0])
    dma_result = reduce(lambda x, y: (1 - y[1]) * x + y[1] * y[0], zip(close, weighted), close.iloc[0])
    return dma_result
```

```python
#@title MCDX
def MCDX_indicator(df):

  indicatorScore = 0
  stxt2 = ""

  try:
    # {MCDX}
    RSIBaseBanker       = 50  #"Banker Base"
    RSIPeriodBanker     = 50  #"Banker RSI Period"
    RSIBaseHotMoney     = 30  #"Hot Money RSI Base"
    RSIPeriodHotMoney   = 40  #"Hot Money RSI Period"
    SensitivityBanker   = 1.5, #"Sensitivity Banker"
    SensitivityHotMoney = 0.7, #"Sensitivity Hot Money"

    def rsi_function(df ,sensitivity, rsiPeriod, rsiBase):
      df2 = df.copy()
      df2['TEMP1'] = sensitivity * (pta.rsi(df2['Close'], rsiPeriod) - rsiBase)
      df2['TEMP1'] = df2['TEMP1'].apply(lambda x: min(max(x, 0), 20))
      return df2['TEMP1']*5


    df['rsi_Banker']  = rsi_function(df,SensitivityBanker,   RSIPeriodBanker,   RSIBaseBanker)
    df['rsi_HotMoney'] = rsi_function(df,SensitivityHotMoney, RSIPeriodHotMoney, RSIBaseHotMoney)

    rsi_HotMoney_list = df['rsi_HotMoney'].to_list()
    rsi_Banker_list = df['rsi_Banker'].to_list()

    Retailer_list =[]
    for ix in range(0,len(rsi_HotMoney_list)):
      if rsi_HotMoney_list[ix] !=  100:
        rsi_HotMoney_list[ix] = rsi_HotMoney_list[ix] - rsi_Banker_list[ix]
        Retailer_list.append( 100 - rsi_HotMoney_list[ix] - rsi_Banker_list[ix])

      if rsi_HotMoney_list[ix] ==  100:
        rsi_HotMoney_list[ix] = rsi_HotMoney_list[ix] - rsi_Banker_list[ix]
        Retailer_list.append(0)

    df["遊資籌碼"] = rsi_HotMoney_list
    df["主力籌碼"] = rsi_Banker_list
    df["散戶籌碼"] = Retailer_list

    主力籌碼 = str(round(df["主力籌碼"].iloc[-1],2))

    if df["主力籌碼"].iloc[-1] >0  :
      indicatorScore = 0

      if df["主力籌碼"].iloc[-1] >= df["主力籌碼"].iloc[-2]:
        stxt2 = "🟩📈"+主力籌碼+"主力吸籌區"
      else:
        stxt2 = "🟩📉"+主力籌碼+"主力吸籌區"


    if df["主力籌碼"].iloc[-1] >= 25:
      indicatorScore = 1

      if df["主力籌碼"].iloc[-1] >= df["主力籌碼"].iloc[-2]:
        stxt2 = "🟩📈"+主力籌碼+"主力建倉區"
      else:
        stxt2 = "🟩📉"+主力籌碼+"主力建倉區"


    if df["主力籌碼"].iloc[-1] >= 50:

      if df["主力籌碼"].iloc[-1] >= df["主力籌碼"].iloc[-2]:
        stxt2 = "🟩📈"+主力籌碼+"主力控盤區"
      else:
        stxt2 = "🟩📉"+主力籌碼+"主力控盤區"

    if df["主力籌碼"].iloc[-1] >= 75:

      if df["主力籌碼"].iloc[-1] >= df["主力籌碼"].iloc[-2]:
        stxt2 = "🟩📈"+主力籌碼+"價格拉升區"
      else:
        stxt2 = "🟩📉"+主力籌碼+"價格拉升區區"

    if df["主力籌碼"].iloc[-1] == 0  :
      stxt2 = "🟥零主力籌碼"
      indicatorScore = -1

  except:
    print("資料不完全，無法正確計算主力籌碼")
    stxt2 = "-"
    indicatorScore = 0

  return_list =[stxt2,indicatorScore]


  return return_list
```

```python
#@title 低吸高拋()
def 低吸高拋(tickerDatas):

  tickerData = tickerDatas.copy()
  stxt=""
  indicatorScore=0


  try:
    #低吸-------------------------------------
    tickerData['RSI1'] = pta.rsi(tickerData['Close'],length=6)
    #低吸 = RSI1[-1]>11 and RSI1[-2] < 11
    低吸DATA=[11]*(len(tickerData))

    tickerData['低吸line']=低吸DATA
    tickerData['低吸'] = pta.cross(tickerData['RSI1'],tickerData['低吸line'])
    低吸_last_three_values_greater_than_zero = (tickerData['低吸'].tail(4) > 0).any()


    if 低吸_last_three_values_greater_than_zero:
      stxt = stxt + "" + "🟩低吸"
      indicatorScore = 1

  except:
    print("資料不完全，無法正確計算低吸")



  try:
    #堅決買進-------------------------------------
    tickerData['AMOUNT'] = (  (tickerData['Close']+ tickerData['High'] + tickerData['Low']) /3 )  * tickerData['Volume']
    tickerData['VAR7'] = pta.ema((tickerData['AMOUNT']/tickerData['Volume']/100),length=3)
    tickerData['VAR8'] = pta.ema(tickerData['VAR7'],length=88)
    tickerData['VAR9'] = tickerData['VAR8']*0.85
    tickerData['VAR9'] = tickerData['VAR9'] *0.87

    tickerData['Close2'] = tickerData['Close'].shift(1)

    tickerData['堅決買進1'] = tickerData['Low'] < tickerData['VAR9']
    tickerData['堅決買進2'] = tickerData['Close'] > tickerData['Close2']*1.02
    tickerData['堅決買進'] = tickerData['堅決買進1'] & tickerData['堅決買進2']

    堅決買進_last_three_values_greater_than_zero = (tickerData['堅決買進'].tail(4) > 0).any()


    if 堅決買進_last_three_values_greater_than_zero:
      stxt = stxt + "" + "🟩堅決買進"
      indicatorScore = 1


  except:
    print("資料不完全，無法正確計算堅決買進")




  try:
    #進攻-------------------------------------
    tickerData['VARF']= round( (2*tickerData['Close']+ tickerData['High'] + tickerData['Low'])/4 ,3)
    # 创建 "LL" 列，取前34行的最小值
    tickerData['LL'] = tickerData['Low'].rolling(window=34, min_periods=1).min()
    # 创建 "HH" 列，取前34行的最大值
    tickerData['HH'] = tickerData['High'].rolling(window=34, min_periods=1).max()
    tickerData['VA6'] =  pta.ema( (tickerData['VARF']-tickerData['LL'])/( tickerData['HH']- tickerData['LL'])*100 ,6 )

    # 创建 "H2" 列，为 "HH" 列的前1行数据
    tickerData['VA62'] = tickerData['VA6'].shift(1)
    tickerData['VA7'] =pta.ema( ((0.667 * tickerData['VA62']) + 0.333 * tickerData['VA6']) ,4)


    tickerData['Close2'] = tickerData['Close'].shift(1)
    z=[]
    for i in range(len(tickerData)):
      count = 0
      if i >= 8:
        for i2 in range(8):
          if tickerData["Close"].iloc[i-i2] < tickerData["Close2"].iloc[i-i2]:
            count += 1

      z.append(count)

    tickerData["z"] = z
    tickerData["出擊cnd01"] = (tickerData["z"]/8) > 0.3

    z1 =[]
    for i in range(len(tickerData)):
      count = 0
      if i >= 3:
        for i2 in range(3):
          if tickerData['VA6'].iloc[i-i2] > tickerData['VA7'].iloc[i-i2]:
            count += 1

      if count > 0:
        cnd = True
      else:
        cnd = False

      z1.append(cnd)

    tickerData["出擊cnd02"] = z1

    tickerData['LL120'] = tickerData['Low'].rolling(window=120, min_periods=1).min()

    cnd03_list =[]
    ans=False
    for i in range(len(tickerData)):
      if i >= 1 :
        ans = (tickerData['Low'].iloc[i-1] == tickerData['LL120'].iloc[i])
      else:
        ans=False

      cnd03_list.append(ans)

    tickerData['出擊cnd03'] = cnd03_list

    tickerData['出擊cnd04'] = tickerData['Close'] > tickerData['Open']

    tickerData['進攻'] = tickerData["出擊cnd01"] & tickerData["出擊cnd02"] & tickerData['出擊cnd03'] & tickerData['出擊cnd04']

    進攻_last_three_values_greater_than_zero = (tickerData['進攻'].tail(4) > 0).any()

    if 進攻_last_three_values_greater_than_zero:
      stxt = stxt + "" + "🟩進攻"
      indicatorScore = 1

  except:
    print("資料不完全，無法正確計算進攻")


  try:
    #波段買入
    tickerData["A"]= ( 3*tickerData['Close']+ tickerData['High'] + tickerData['Low'] + tickerData['Open']) / 6
    d1_list = []
    for i2 in range(len(tickerData)):
      d2=0
      d1 = [0]*20
      if i2 >= 20:
        for i in range(20):
          #print('波段買入1 = 波段買入1 + '+ str( (21-i) ) + '* tickerData["A"]['+str(i2-i)+']')
          d1[i-1] = (20-i)* tickerData["A"].iloc[i2-i]

        d2 = round( sum(d1) / 211 ,3)

      d1_list.append(d2)

    tickerData["d1"] = d1_list
    tickerData['d2'] = pta.ema(tickerData["d1"],length=2)
    tickerData['d3'] = pta.ema(tickerData["d2"],length=2)
    tickerData['k1'] = pta.ema(tickerData["d3"],length=2)

    tickerData['波段買入'] = pta.cross(tickerData["d1"],tickerData['k1'])
    波段買入_last_three_values_greater_than_zero = (tickerData['波段買入'].tail(2) > 0).any()

    if 波段買入_last_three_values_greater_than_zero:
      stxt = "🟩波段買入"
      indicatorScore = 1

  except:
    print("資料不完全，無法正確計算波段買入")


  try:
    #高拋------------------------
    tickerData['LL27'] = tickerData['Low'].rolling(window=27, min_periods=1).min()
    tickerData['HH27'] = tickerData['High'].rolling(window=27, min_periods=1).max()
    tickerData['A00'] = pta.ema(   (tickerData['Close'] - tickerData['LL27']) / (tickerData['HH27']- tickerData['LL27'])*30,4,append=True)
    tickerData['A01'] = tickerData['A00'] * 7
    tickerData['A020'] = pta.ema(tickerData['A00'],3,append=True)
    tickerData['A021'] = tickerData['A020']*3
    tickerData['A03'] = pta.ema( tickerData['A020'] ,2,append=True)

    tickerData['多方趨勢'] = tickerData['A01'] - tickerData['A021'] - tickerData['A03']
    天線=[85]*(len(tickerData))

    tickerData['天線']=天線
    tickerData['頂'] = pta.cross(tickerData['天線'],tickerData['多方趨勢'])
    # 检查 頂" 列的最后三个值是否大于0
    頂_last_three_values_greater_than_zero = (tickerData['頂'].tail(3) > 0).any()

    if 頂_last_three_values_greater_than_zero:
      stxt = stxt + "" + "🟥高拋"
      indicatorScore = -1


  except:
    print("資料不完全，無法正確計算高拋")












  return_list = [stxt,indicatorScore]
  return return_list
```

```python
#@title 相對PE、PB與 PS 估價 Valuation
def Valuation(TickerName):
  # 相對PE、PB與 PS 估價

  #透過 fake-useragent 自動生成假的 header 解決被網頁認為是機器人的問題
  # 使用假header
  ua = UserAgent()
  # 隨機產生的 UA
  user_agent = ua.random
  headers = {'User-Agent': user_agent}
  # 連結
  TickerName = TickerName.lower()
  url = r'https://stockanalysis.com/stocks/'+TickerName+'/financials/ratios/'

  PE = ['','','']
  PS = ['','','']
  PB = ['','','']
  try:
    request = requests.get(url, headers = headers)
    request.encoding = 'utf-8'
    #data = pd.read_html(request.text)[0]
    stri = StringIO(request.text)
    data = pd.read_html(stri)[0]


    PE_Ratio_Current = float(data.iloc[4,1])
    PS_Ratio_Current = float(data.iloc[6,1])
    PB_Ratio_Current = float(data.iloc[7,1])

    PE[1] = PE_Ratio_Current
    PB[1] = PB_Ratio_Current
    PS[1] = PS_Ratio_Current


  except:
    #print('無法抓取資料')
    pass

  now = datetime.datetime.now()
  noeyear = now.strftime('%Y')

  pe_list=[]
  pb_list=[]
  ps_list=[]
  年 = 5

  perror = []
  for i in range(1,年+1):
    delta = datetime.timedelta(days=365*i)
    d = (now - delta)
    d1 = str(d.year)

    try:
      pe1 = data.iloc[4,i]
      pe = float(pe1)
      pe_list.append(pe)

      ps1 = data.iloc[6,i]
      ps = float(ps1)
      ps_list.append(ps)


      pb1 = data.iloc[7,i]
      pb = float(pb1)
      pb_list.append(pb)


    except:
      exceptTXT = "無法取得"+d1+'年資料'
      perror.append("True")
      #print(exceptTXT2)


  if not(("True" in perror)):

    PE_Ratio_avg = round(np.mean(pe_list),2)
    PB_Ratio_avg = round(np.mean(pb_list),2)
    PS_Ratio_avg = round(np.mean(ps_list),2)

    PE[2] = PE_Ratio_avg
    PS[2] = PS_Ratio_avg
    PB[2] = PB_Ratio_avg



    if PE_Ratio_Current < PE_Ratio_avg :
      PE[0] = "🟩"
    else:
      if PE_Ratio_Current == PE_Ratio_avg :
        PE[0] = "🟨"
      else:
        PE[0] = "🟥"


    if PS_Ratio_Current < PS_Ratio_avg :
      PS[0] = "🟩"
    else:
      if PS_Ratio_Current == PS_Ratio_avg :
        PS[0] = "🟨"
      else:
        PS[0] = "🟥"


    if PB_Ratio_Current < PB_Ratio_avg :
      PB[0] = "🟩"
    else:
      if PB_Ratio_Current == PB_Ratio_avg :
        PB[0] = "🟨"
      else:
        PB[0] = "🟥"




  return PE+PB+PS
```

```python
#@title 30日歴史波動率
def GetHV2(df):

  #ticker = yf.Ticker(TickerName)
  #Data = ticker.history(period = '30D',auto_adjust=True)
  close_list = df['Close'].tail(30).to_list()
  HV_list=[0]
  listLen = len(close_list)

  i = 0
  for close0 in close_list:

    if i < listLen-1:
      HV_list.append(((close_list[i+1] - close0)/close_list[i+1])*100)
      i=i+1


  HV = round(np.std(HV_list) * (250**0.5),2)
  return HV
```

```python
#@title 計算連續上漲與連續下跌天數
def up_days(tickerData):
  # 計算連續上漲與連續下跌天數
  close_prices = tickerData["Close"]
  continuous_up_days = 0
  continuous_down_days = 0
  previous_close = None

  days_list = []

  for close in close_prices:
      if previous_close is not None:
          if close > previous_close:
              continuous_up_days += 1
              continuous_down_days = 0
          elif close < previous_close:
              continuous_down_days += 1
              continuous_up_days = 0
          else:
              continuous_up_days = 0
              continuous_down_days = 0
      days_list.append(max(continuous_up_days, continuous_down_days*-1))
      previous_close = close

  # 将 "days" 列添加到原始DataFrame 中
  tickerData["days"] = days_list
```

```python
#@title CYC 無窮成本均線
def CYC_indicator(df,TickerName):

  TickerName = TickerName.upper()
  '''
  # 取得股票的資料
  ticker = yf.Ticker(TickerName)
  tickerData = ticker.history(period = '2Y',auto_adjust=False)
  '''
  tickerData = df.copy()
  try:
    # 取得股票的資訊
    data = yf.Ticker(TickerName).info
    # 取得流通股
    shares_outstanding = data["sharesOutstanding"]
  except:
    #print('取得流通股異常')
    shares_outstanding = 0


  #tickerData['amount'] = tickerData["Close"] * tickerData['Volume']

  tickerData['amount'] = ((tickerData['High']+tickerData['Low']+tickerData["Close"])/3) * tickerData['Volume']
  ticker_Volume = tickerData['Volume']

  JJJT_list = []
  for i in range(0, len(ticker_Volume)):
    jjj = 0
    #if ticker_Volume[i] > 0.01 :
    if ticker_Volume.iloc[i] > 0.01 :
      jjj = 0.01*tickerData['amount'].iloc[i]/tickerData['Volume'].iloc[i]
    else:
      jjj = tickerData["Close"].iloc[i]

    ddd = (tickerData['High'].iloc[i] < 0.01 or tickerData['Low'].iloc[i] < 0.01 )

    if ddd :
      JJJT = True
    else:
      JJJT = ( jjj < tickerData['High'].iloc[i]+0.01 ) and  ( jjj > tickerData['Low'].iloc[i] - 0.01 )

    JJJT_list.append(JJJT)

  tickerData['TTTJ'] = JJJT_list

  p=5
  cyc5_1_list = ( (tickerData['amount'].ewm(span=p).mean() ) /  tickerData['Volume'].ewm(span=p).mean() )
  cyc5_2_list = ((tickerData['High']+tickerData['Low']+tickerData['Close'])/3).ewm(span=p).mean()

  p=13
  cyc13_1_list = ( (tickerData['amount'].ewm(span=p).mean() ) /  tickerData['Volume'].ewm(span=p).mean() )
  cyc13_2_list = ((tickerData['High']+tickerData['Low']+tickerData['Close'])/3).ewm(span=p).mean()


  p=34
  cyc34_1_list = ( (tickerData['amount'].ewm(span=p).mean() ) /  tickerData['Volume'].ewm(span=p).mean() )
  cyc34_2_list = ((tickerData['High']+tickerData['Low']+tickerData['Close'])/3).ewm(span=p).mean()


  cyc88_1_list = [0]

  p=120
  cyc88_2_list = ((tickerData['High']+tickerData['Low']+tickerData['Close'])/3).ewm(span=p).mean()

  for i in range(1, len(ticker_Volume)):
    if shares_outstanding > 0:
      #數據
      data_prices = tickerData['amount'].head(i)/( (100 * tickerData['Volume']).head(i) )

      #權重
      weighted = (100 * tickerData['Volume'].head(i) / shares_outstanding / 100)

      #調用函數計算動態移動平均線
      dma_result = DMA(data_prices,weighted)
    else:
      dma_result = cyc88_2_list


    cyc88_1_list.append(dma_result)

  p=120
  cyc88_2_list = ((tickerData['High']+tickerData['Low']+tickerData['Close'])/3).ewm(span=p).mean()

  cyc5_3_list=[]
  cyc13_3_list=[]
  cyc34_3_list=[]
  cyc88_3_list=[]

  for i in range(0, len(ticker_Volume)):
    if JJJT_list[i] :
      #cyc5_3_list.append(cyc5_1_list[i])
      cyc5_3_list.append(cyc5_1_list.iloc[i])
      #cyc13_3_list.append(cyc13_1_list[i])
      cyc13_3_list.append(cyc13_1_list.iloc[i])
      #cyc34_3_list.append(cyc34_1_list[i])
      cyc34_3_list.append(cyc34_1_list.iloc[i])
      #cyc88_3_list.append(cyc88_1_list[i])
      cyc88_3_list.append(cyc88_1_list.iloc[i])
    else:
      #cyc5_3_list.append(cyc5_2_list[i])
      cyc5_3_list.append(cyc5_2_list.iloc[i])
      #cyc13_3_list.append(cyc13_2_list[i])
      cyc13_3_list.append(cyc13_2_list.iloc[i])
      #cyc34_3_list.append(cyc34_2_list[i])
      cyc34_3_list.append(cyc34_2_list.iloc[i])
      #cyc88_3_list.append(cyc88_2_list[i])
      cyc88_3_list.append(cyc88_2_list.iloc[i])

  tickerData['CYC5']=cyc5_3_list
  tickerData['CYC13']=cyc13_3_list
  tickerData['CYC34']=cyc34_3_list
  tickerData['CYC∞']=cyc88_3_list


  tickerData['GD'] = tickerData['CYC5'] > tickerData['CYC13']
  tickerData['DD'] = tickerData['CYC5'] < tickerData['CYC13']
  tickerData['CYC∞U'] = tickerData['Close'] > tickerData['CYC∞']

  tickerData['Tangled'] = ( (tickerData['CYC5'] < tickerData['High']) &
                (tickerData['CYC5'] > tickerData['Low']) &
                (tickerData['CYC13'] < tickerData['High']) &
                (tickerData['CYC13'] > tickerData['Low']) &
                (tickerData['CYC34'] < tickerData['High']) &
                (tickerData['CYC34'] > tickerData['Low']) &
                (tickerData['CYC∞'] < tickerData['High']) &
                (tickerData['CYC∞'] > tickerData['Low'])

                )



  GD_list = tickerData['GD'].iloc[-20:].to_list()
  DD_list = tickerData['DD'].iloc[-20:].to_list()
  U_list = tickerData['CYC∞U'].iloc[-20:].to_list()



  reversed_GD_list = GD_list[::-1]
  reversed_DD_list = DD_list[::-1]
  reversed_U_list = U_list[::-1]


  if any(reversed_GD_list):
    first_true_index = reversed_GD_list.index(True)

  first_true_index = 0

  #計算金叉天數
  GD_days = 0
  if first_true_index == 0:
    for g in reversed_GD_list:
      if g :
        GD_days +=1
      else:
        break
  else:
    if first_true_index < 3 :
      GD_days = first_true_index
      for i in range(first_true_index,len(reversed_GD_list)-1):
        if reversed_GD_list[i] :
          GD_days +=1
        else:
          break


  if GD_days == 1 and reversed_GD_list[0]:
    GD_days = GD_days
  else:
    GD_days  = 0


  if GD_days == 1:
    if reversed_GD_list[0]:
      GD_days = GD_days
    else:
      GD_days  = 0
  else:
    GD_days = GD_days







  if GD_days > 0 and GD_days <= 5 :
    stxt1="💎"+str(GD_days)
    indicator_list1 =1
  else:
    #if (tickerData['CYC5'][-1] < tickerData['CYC13'][-1]) and (tickerData['CYC5'][-2] > tickerData['CYC13'][-2])  :
    if (tickerData['CYC5'].iloc[-1] < tickerData['CYC13'].iloc[-1]) and (tickerData['CYC5'].iloc[-2] > tickerData['CYC13'].iloc[-2])  :
      stxt1="☠️"
      indicator_list1 =-1
    else:
      stxt1=""
      indicator_list1 = 0

  #計算死叉天數
  DD_days = 0
  for D in reversed_DD_list:
    if D :
      DD_days+=1
    else:
      break

  if DD_days == 1:
    if reversed_DD_list[0]:
      DD_days = DD_days
    else:
      DD_days  = 0

  DD_days_01 = DD_days



  #計算站上無窮天數
  U_days = 0
  for u in reversed_U_list:
    if u :
      U_days +=1
    else:
      break

  if U_days > 20:
    U_days = 0

  if GD_days > 5:
    GD_days = 0

  U_days_max = GD_days + 6

  if U_days > 0 and U_days < U_days_max :
    stxt2="🟩" + str(U_days)
    indicator_list2 =1
  else:
    #if (tickerData['Close'][-1] < tickerData['CYC∞'][-1]) and (tickerData['Close'][-2] > tickerData['CYC∞'][-2]):
    if (tickerData['Close'].iloc[-1] < tickerData['CYC∞'].iloc[-1]) and (tickerData['Close'].iloc[-2] > tickerData['CYC∞'].iloc[-2]):
      stxt2="🟥"
      indicator_list2 =-1
    else:
      stxt2=""
      indicator_list2 =0


  #計算糾結天數
  #取得金叉前數據
  #Tangled_len =  (  len(tickerData['Close']) -  GD_days - 4  ) #取得金叉前5天
  Tangled_len =  (  len(tickerData['Close']) )

  first_Tangled_rows = tickerData.head(Tangled_len)
  Tangled_rows = first_Tangled_rows.tail(120)
  Tangled_list = Tangled_rows['Tangled'].to_list()


  reversedTangled_list = Tangled_list[::-1]
  T_days = 0
  if any(reversedTangled_list):
    for T in reversedTangled_list:
      if not T :
        T_days +=1
      else:
        break

    if T_days == 1:
      if reversedTangled_list[0]:
        T_days = T_days
      else:
        T_days  = 0

    if T_days > 0 :
      if DD_days == 0:
        stxt3 = "🌈" + str(T_days)
      else:
        if DD_days > T_days :
          stxt3 = "🌈" + str(T_days)
        else:
          stxt3 = ""
    else:
      stxt3 = ""
  else:
    stxt3 = ""

  '''
  if T_days > 0 and ( U_days > 0 or GD_days > 0 ) and (T_days < DD_days):
    T_days = T_days + 5 + GD_days
    stxt3 = "🌈" + str(T_days)
  else:
    stxt3 = ""
  '''

  return stxt1,stxt2,stxt3,indicator_list1,indicator_list2
```

```python
#@title 均線crossover
def Crossover_indicator(df):
  tickerData = df.copy()
  tickerData['Line11'] = tickerData['Line1'].shift(1)
  tickerData['Line22'] = tickerData['Line2'].shift(1)


  if( ((tickerData['Line1'].iloc[-1] > tickerData['Line2'].iloc[-1]) and (tickerData['Line11'].iloc[-1] < tickerData['Line22'].iloc[-1])) or
    ((tickerData['Line1'].iloc[-2] > tickerData['Line2'].iloc[-2]) and (tickerData['Line11'].iloc[-2] < tickerData['Line22'].iloc[-2])) or
    ((tickerData['Line1'].iloc[-3] > tickerData['Line2'].iloc[-3]) and (tickerData['Line11'].iloc[-3] < tickerData['Line22'].iloc[-3])) ) :
    stext = "G🟩"
    indicator = 1
  elif( ((tickerData['Line1'].iloc[-1] < tickerData['Line2'].iloc[-1]) and (tickerData['Line11'].iloc[-1] > tickerData['Line22'].iloc[-1])) or
    ((tickerData['Line1'].iloc[-2] < tickerData['Line2'].iloc[-2]) and (tickerData['Line11'].iloc[-2] > tickerData['Line22'].iloc[-2])) or
    ((tickerData['Line1'].iloc[-3] < tickerData['Line2'].iloc[-3]) and (tickerData['Line11'].iloc[-3] > tickerData['Line22'].iloc[-3])) ) :
    stext = "D🟥"
    indicator = -1
  elif tickerData['Line1'].iloc[-1] > tickerData['Line2'].iloc[-1]:
    stext = "🟩"
    indicator = 1
  else:
    stext = "🟥"
    indicator = -1

  return stext,indicator
```

```python
#@title 六脈神劍 SIXMDS_indicator

def SIXMDS_indicator(tickerData):


  #01 --- macd --- #
  # 設置參數
  EMA_SHORT = 8  # MACD 指標的短期 EMA
  EMA_LONG = 13   # MACD 指標的長期 EMA
  SIGNAL = 5      # MACD 指標信號線的 EMA
  try:
    # 計算 MACD 指標
    tickerData['EMA_short'] = tickerData['Close'].ewm(span=EMA_SHORT, adjust=False).mean()
    tickerData['EMA_long'] = tickerData['Close'].ewm(span=EMA_LONG, adjust=False).mean()
    tickerData['diff'] = tickerData['EMA_short'] - tickerData['EMA_long']
    tickerData['dea'] = tickerData['diff'].ewm(span=SIGNAL, adjust=False).mean()
    #tickerData['Histogram'] = tickerData['diff'] - tickerData['dea']
  except:
    print("MACD 無法正確計算")
    tickerData['diff'] = 0
    tickerData['dea'] = 0




  #02 --- KDJ --- #
  # 設置參數
  N_KDJ = 8  # KDJ 指標的窗口長度
  # 計算 KDJ 指標

  try:
    tickerData['Low_N'] = tickerData['Low'].rolling(window=N_KDJ).min()
    tickerData['High_N'] = tickerData['High'].rolling(window=N_KDJ).max()
    tickerData['RSV'] = ((tickerData['Close'] - tickerData['Low_N']) / (tickerData['High_N'] - tickerData['Low_N']) )* 100

    # 初始化 K 和 D 值
    tickerData['K'] = tickerData['RSV'].ewm(alpha=1/3).mean()
    tickerData['D'] = tickerData['K'].ewm(alpha=1/3).mean()
    #tickerData['J'] = 3 * tickerData['K'] - 2 * tickerData['D']
  except:
    print("KDJ 無法正確計算")
    tickerData['K'] = 0
    tickerData['D'] = 0

  #03 --- RSI --- #
  try:
    tickerData['RSI1'] = pta.rsi(tickerData['Close'],5)
    tickerData['RSI2'] = pta.rsi(tickerData['Close'],13)
  except:
    print("RSI 無法正確計算")
    tickerData['RSI1'] = 0
    tickerData['RSI2'] = 0



  #04 --- LWR --- #

  try:

    # 計算最近 13 天的最高價和最低價
    n=13
    tickerData['High_N'] = tickerData['High'].rolling(window=13).max()
    tickerData['Low_N'] = tickerData['Low'].rolling(window=13).min()
    #計算 HHV 和 LLV
    #tickerData['hhv']= tickerData['High_N'].rolling(window=n0.max()
    #tickerData['llv'] = tickerData['Low_N'].rolling(window=n).min()

    # 計算 LWR
    #tickerData['LWR'] = -(tickerData['hhv'] - tickerData['Close']) / (tickerData['hhv'] - tickerData['llv']) * 100




    # 計算 LWR 指標
    tickerData['LWR'] = -( (tickerData['High_N'] - tickerData['Close']) / (tickerData['High_N'] - tickerData['Low_N']) )* 100
    tickerData['LWR1'] = pta.sma(tickerData['LWR'],3,1)
    tickerData['LWR2'] = pta.sma(tickerData['LWR1'],3,1)

    #tickerData['LWR1'] = tickerData['LWR'].rolling(window=13).mean()
    #tickerData['LWR2'] = tickerData['LWR1'].rolling(window=3).mean()

  except:
    print("LWR 無法正確計算")
    tickerData['LWR'] = 0
    tickerData['LWR1'] = 0
    tickerData['LWR2'] = 0


  #05 ----  BBI  --- #
  #tickerData['BBI'] = (pta.ma(tickerData['Close'],3)+pta.sma(tickerData['Close'],6)+pta.sma(tickerData['Close'],12)+pta.sma(tickerData['Close'],24))/4

  tickerData['BBI'] = ((tickerData['Close'].rolling(window=3).mean())+(tickerData['Close'].rolling(window=6).mean())+(tickerData['Close'].rolling(window=12).mean())+(tickerData['Close'].rolling(window=24).mean()))/4

  #06 ----  MTM  --- #
  # 設置計算 MTM 的周期 n
  n = 1  # 通常為 12（可以根據需求調整）
  # 計算 MTM 指標
  try:
    tickerData['MTM'] = tickerData['Close'] - tickerData['Close'].shift(n)
    tickerData['MMS'] = (pta.ema(pta.ema(tickerData['MTM'],5),3)/pta.ema(pta.ema(abs(tickerData['MTM']),5),3))*100
    tickerData['MMM'] = (pta.ema(pta.ema(tickerData['MTM'],13),8)/pta.ema(pta.ema(abs(tickerData['MTM']),13),8))*100
  except:
    print("MTM 無法正確計算")
    tickerData['MTM'] = 0
    tickerData['MMS'] = 0
    tickerData['MMM'] = 0




  tickerData['six1'] = (tickerData['diff'] > tickerData['dea']).astype(int)
  tickerData['six2'] = (tickerData['K'] > tickerData['D']).astype(int)
  tickerData['six3'] = (tickerData['RSI1'] > tickerData['RSI2']).astype(int)
  tickerData['six4'] = (tickerData['LWR1'] > tickerData['LWR2']).astype(int)
  tickerData['six5'] = (tickerData['Close'] > tickerData['BBI']).astype(int)
  tickerData['six6'] = (tickerData['MMS'] > tickerData['MMM']).astype(int)

  SIXMDS=tickerData['six1'].iloc[-1]+tickerData['six2'].iloc[-1]+tickerData['six3'].iloc[-1]+tickerData['six4'].iloc[-1]+tickerData['six5'].iloc[-1]+tickerData['six6'].iloc[-1]


  if SIXMDS > 0 :
    stxt1= str(SIXMDS)+"🟩"+str(6-SIXMDS)+"🟥"
  else:
    stxt1= str(6-SIXMDS)+"🟥"


  SIXMDS_list=[stxt1,SIXMDS,tickerData['six1'].iloc[-1],tickerData['six2'].iloc[-1],tickerData['six3'].iloc[-1],tickerData['six4'].iloc[-1],tickerData['six5'].iloc[-1],tickerData['six6'].iloc[-1]]

  if SIXMDS == 6 :
    indicator_score = 1
  elif SIXMDS <= 5 :
    indicator_score = 0
  elif SIXMDS <= 3:
    indicator_score = _1

  return SIXMDS_list,indicator_score
```

```python
#@title supertrend 指標
def supertrend(df, period=10, atr_multiplier=3):
  stext=""
  df2 = df.copy()
  def tr(data):
      data['previous_close'] = data['Close'].shift(1)
      data['high-low'] = abs(data['High'] - data['Low'])
      data['high-pc'] = abs(data['High'] - data['previous_close'])
      data['low-pc'] = abs(data['Low'] - data['previous_close'])

      tr = data[['high-low', 'high-pc', 'low-pc']].max(axis=1)

      return tr

  def atr(data, period):
      data['tr'] = tr(data)
      atr = data['tr'].rolling(period).mean()
      return atr


  hl2 = (df['High'] + df['Low']) / 2
  df['atr'] = atr(df, period)
  df['upperband'] = hl2 + (atr_multiplier * df['atr'])
  df['lowerband'] = hl2 - (atr_multiplier * df['atr'])
  df['in_uptrend'] = True

  for current in range(1, len(df.index)):
    previous = current - 1
    if df['Close'].iloc[current] > df['upperband'].iloc[previous]:
              df.loc[df.index[current], 'in_uptrend'] = True  # 使用 .loc
    elif df['Close'].iloc[current] < df['lowerband'].iloc[previous]:
        df.loc[df.index[current], 'in_uptrend'] = False  # 使用 .loc
    else:
        df.loc[df.index[current], 'in_uptrend'] = df['in_uptrend'].iloc[previous]  # 使用 .loc

        if df['in_uptrend'].iloc[current] and df['lowerband'].iloc[current] < df['lowerband'].iloc[previous]:
            df.loc[df.index[current], 'lowerband'] = df['lowerband'].iloc[previous]  # 使用 .loc

        if not df['in_uptrend'].iloc[current] and df['upperband'].iloc[current] > df['upperband'].iloc[previous]:
            df.loc[df.index[current], 'upperband'] = df['upperband'].iloc[previous]  # 使用 .loc

  df['in_uptrend2'] = df['in_uptrend'].shift(1)
  df['supertrend_buy']  = (df['in_uptrend'] == True) & (df['in_uptrend2'] == False )
  df['supertrend_sell'] = (df['in_uptrend'] == False) & (df['in_uptrend2'] == True)

  if  df['supertrend_buy'].iloc[-1] or df['supertrend_buy'].iloc[-2] or df['supertrend_buy'].iloc[-3]:
    stext = "Buy🟩"
    indicator = 1
  elif df['supertrend_sell'].iloc[-1] :
    stext = "Sell🟥"
    indicator = -1
  elif df['in_uptrend'].iloc[-1] == True :
    stext = "🟩"
    indicator = 1
  else:
    stext = "🟥"
    indicator = -1




  return stext,indicator
```

```python
#@title TRENDEXPERT 趨勢王指標
def TRENDEXPERT(df):
  CLOSE = df.Close
  LOW = df.Low
  HIGH = df.High
  OPEN = df.Open
  A1=(MyTT.EMA((OPEN+HIGH+LOW+CLOSE)/4,3)+MyTT.EMA((OPEN+HIGH+LOW+CLOSE)/4,6)+MyTT.EMA((OPEN+HIGH+LOW+CLOSE)/4,9))/3
  A2=(MyTT.EMA((OPEN+HIGH+LOW+CLOSE)/4,5)+MyTT.EMA((OPEN+HIGH+LOW+CLOSE)/4,10)+MyTT.EMA((OPEN+HIGH+LOW+CLOSE)/4,20))/3
  A3=(MyTT.EMA((OPEN+HIGH+LOW+CLOSE)/4,7)+MyTT.EMA((OPEN+HIGH+LOW+CLOSE)/4,14)+MyTT.EMA((OPEN+HIGH+LOW+CLOSE)/4,28))/3
  A4=(MyTT.EMA((OPEN+HIGH+LOW+CLOSE)/4,9)+MyTT.EMA((OPEN+HIGH+LOW+CLOSE)/4,18)+MyTT.EMA((OPEN+HIGH+LOW+CLOSE)/4,36))/3
  A5=(MyTT.EMA((OPEN+HIGH+LOW+CLOSE)/4,11)+MyTT.EMA((OPEN+HIGH+LOW+CLOSE)/4,22)+MyTT.EMA((OPEN+HIGH+LOW+CLOSE)/4,44))/3
  A6=(MyTT.EMA((OPEN+HIGH+LOW+CLOSE)/4,13)+MyTT.EMA((OPEN+HIGH+LOW+CLOSE)/4,26)+MyTT.EMA((OPEN+HIGH+LOW+CLOSE)/4,52))/3
  A7=(MyTT.EMA((OPEN+HIGH+LOW+CLOSE)/4,21)+MyTT.EMA((OPEN+HIGH+LOW+CLOSE)/4,34)+MyTT.EMA((OPEN+HIGH+LOW+CLOSE)/4,68))/3

  VAR1=MyTT.FORCAST(A1,6)
  VAR2=MyTT.FORCAST(A2,6)
  VAR3=MyTT.FORCAST(A3,6)
  VAR4=MyTT.FORCAST(A4,6)
  VAR5=MyTT.FORCAST(A5,6)
  VAR6=MyTT.FORCAST(A6,6)
  VAR7=MyTT.FORCAST(A7,6)

  TG01 =	MyTT.IF(VAR1>REF(VAR1,1),1,0)
  TG02 =	MyTT.IF(VAR2>REF(VAR2,1),1,0)
  TG03 =	MyTT.IF(VAR3>REF(VAR3,1),1,0)
  TG04 =	MyTT.IF(VAR4>REF(VAR4,1),1,0)
  TG05 =	MyTT.IF(VAR5>REF(VAR5,1),1,0)
  TG06 =	MyTT.IF(VAR6>REF(VAR6,1),1,0)
  TG07 =  MyTT.IF(VAR7>REF(VAR7,1),1,0)

  TR01 =	MyTT.IF(VAR1<REF(VAR1,1),1,0)
  TR02 =	MyTT.IF(VAR2<REF(VAR2,1),1,0)
  TR03 =	MyTT.IF(VAR3<REF(VAR3,1),1,0)
  TR04 =	MyTT.IF(VAR4<REF(VAR4,1),1,0)
  TR05 =	MyTT.IF(VAR5<REF(VAR5,1),1,0)
  TR06 =	MyTT.IF(VAR6<REF(VAR6,1),1,0)
  TR07 =	MyTT.IF(VAR7<REF(VAR7,1),1,0)

  TG08 = TG01+TG02+TG03+TG04+TG05+TG06+TG07
  TR08 = TR01+TR02+TR03+TR04+TR05+TR06+TR07

  TGC =	MyTT.IF(TG08==7,True,False)
  TRC =	MyTT.IF(TR08==7,True,False)

  df["TGD"] = TG08
  df["TRD"] = TR08
  df["TGC"] = TGC
  df["TRC"] = TRC


  if  df["TGC"].iloc[-1] or df["TGC"].iloc[-2] or df["TGC"].iloc[-3]:
    stext = "🟩"
    indicator = 1
  elif df["TRC"].iloc[-1] or df["TGC"].iloc[-2] or df["TGC"].iloc[-3]:
    stext = "🟥"
    indicator = -1
  else:
    stext = ""
    indicator = 0


  return stext,indicator
```

```python
#@title 金十字考夫曼的自適應移動平均線 GC_KAMA()

def GC_KAMA(tickerData):

  # 收盤價作為 price
  tickerData['price'] = tickerData['Close']

  # 使用者參數
  len1 = 12  # Fast
  len2 = 26  # Slow
  fast = 2   # Fastest SC
  slow = 30  # Slowest SC

  # 計算平滑常數
  fastestSC = 2 / (fast + 1)
  slowestSC = 2 / (slow + 1)

  # 第一組計算
  tickerData['change1'] = (tickerData['price'] - tickerData['price'].shift(len1)).abs()
  tickerData['diff1'] = (tickerData['price'] - tickerData['price'].shift(1)).abs()
  tickerData['volatility1'] = tickerData['diff1'].rolling(window=len1).sum()
  tickerData['ER1'] = tickerData['change1'] / tickerData['volatility1']
  tickerData['SC1'] = (tickerData['ER1'] * (fastestSC - slowestSC) + slowestSC) ** 2

  # 初始化並更新 out1
  out1 = [np.nan] * len(tickerData)  # 初始化 out1
  for i in range(len(tickerData)):
      if i == 0 or np.isnan(out1[i - 1]):
          out1[i] = tickerData['price'].iloc[i]
      else:
          prev_out1 = out1[i - 1]
          price = tickerData['price'].iloc[i]
          SC1 = tickerData['SC1'].iloc[i]
          out1[i] = prev_out1 + SC1 * (price - prev_out1)
  tickerData['out1'] = out1

  # 第二組計算
  tickerData['change2'] = (tickerData['price'] - tickerData['price'].shift(len2)).abs()
  tickerData['diff2'] = (tickerData['price'] - tickerData['price'].shift(1)).abs()
  tickerData['volatility2'] = tickerData['diff2'].rolling(window=len2).sum()
  tickerData['ER2'] = tickerData['change2'] / tickerData['volatility2']
  tickerData['SC2'] = (tickerData['ER2'] * (fastestSC - slowestSC) + slowestSC) ** 2

  # 初始化並更新 out2
  out2 = [np.nan] * len(tickerData)  # 初始化 out2
  for i in range(len(tickerData)):
      if i == 0 or np.isnan(out2[i - 1]):
          out2[i] = tickerData['price'].iloc[i]
      else:
          prev_out2 = out2[i - 1]
          price = tickerData['price'].iloc[i]
          SC2 = tickerData['SC2'].iloc[i]
          out2[i] = prev_out2 + SC2 * (price - prev_out2)
  tickerData['out2'] = out2

  tickerData['GC_KAMA']= (tickerData['out1'] >= tickerData['out2']) & (tickerData['price'] >= tickerData['out1'])


  GC_KAMA = tickerData['GC_KAMA'].to_list()
  GC_KAMA.reverse()

  GC_KAMA_list = [1 if x else 0 for x in GC_KAMA]

  # 找出第 1 個 0 的位置
  first_zero_index = GC_KAMA_list.index(0)

  # 保留第 n 筆之前的資料（含第 5 筆）
  GC_KAMA_list2 = GC_KAMA_list[:first_zero_index]

  days = sum(GC_KAMA_list2)

  if GC_KAMA_list[0] == 1:
    stext = "🟩"+str(days)
    indicator = 1
  else:
    stext = "🟥"
    indicator = -1


  return stext,indicator
```

```python
#@title 考夫曼自適應移動平均線 KAMA()
def KAMA(tickerData):
  # 基本參數
  N1 = 5
  N2 = 50
  N01 = 10
  G = 2  # 增益 (Gain)

  # 計算平滑常數
  FAST = G / (N1 + 1)
  SLOW = G / (N2 + 1)

  # 計算 DIR2 和 VIR2
  tickerData['DIR2'] = (tickerData['Close'] - tickerData['Close'].shift(N01)).abs()
  tickerData['VIR2'] = tickerData['Close'].diff().abs().rolling(window=N01).sum()

  # 效率比 ER2
  tickerData['ER2'] = tickerData['DIR2'] / tickerData['VIR2']

  # 平滑常數 CS2 和 CQ2
  tickerData['CS2'] = tickerData['ER2'] * (FAST - SLOW) + SLOW
  tickerData['CQ2'] = tickerData['CS2'] ** 2

  # 計算 KAMA (模擬 DMA)
  kama = [np.nan] * len(tickerData)  # 初始化 KAMA
  for i in range(len(tickerData)):
      if i == 0 or np.isnan(kama[i - 1]):
          kama[i] = tickerData['Close'].iloc[i]
      else:
          kama[i] = kama[i - 1] + tickerData['CQ2'].iloc[i] * (tickerData['Close'].iloc[i] - kama[i - 1])
  tickerData['KAMA'] = kama

  # 牛熊分界
  #tickerData['BullBearBoundary'] = tickerData['KAMA']

  # KAMA2 條件
  #tickerData['KAMA2'] = np.where(tickerData['KAMA'] / tickerData['KAMA'].shift(1) >= 1,tickerData['KAMA'],np.nan)

  tickerData['KAMA2'] = tickerData['KAMA'] >= tickerData['KAMA'].shift(1)


  KAMA2 = tickerData['KAMA2'].to_list()
  KAMA2.reverse()


  KAMA_list = [1 if x else 0 for x in KAMA2]

  # 找出第 1 個 0 的位置
  first_zero_index = KAMA_list.index(0)

  # 保留第 n 筆之前的資料（含第 5 筆）
  KAMA_list2 = KAMA_list[:first_zero_index]

  days = sum(KAMA_list2)

  if KAMA_list[0] == 1:
    stext = "🟩"+str(days)
    indicator = 1
  else:
    stext = "🟥"
    indicator = 0


  return stext,indicator
```
